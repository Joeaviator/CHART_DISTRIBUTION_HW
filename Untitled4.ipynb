```python

# Check the distribution of all the quantitative features using boxplot
df.boxplot()
plt.title('Boxplot of Quantitative Features')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
df.boxplot()
```

```python
quantitative_cols = [
    'ENGINESIZE',
    'CYLINDERS',
    'FUELCONSUMPTION_CITY',
    'FUELCONSUMPTION_HWY',
    'FUELCONSUMPTION_COMB',
    'FUELCONSUMPTION_COMB_MPG',
    'CO2EMISSIONS'
]

for col in quantitative_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='FUELTYPE', y=col, data=df, palette='viridis', hue='FUELTYPE', legend=False)
    plt.title(f'Distribution of {col} by FUELTYPE')
    plt.xlabel('Fuel Type')
    plt.ylabel(col)
    plt.show()
```

```python
sns.boxplot(x='VEHICLECLASS', y='CO2EMISSIONS', data=df, palette='Spectral', hue='VEHICLECLASS', legend=False)
plt.title('Distribution of CO2 Emissions by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
sns.violinplot(x='FUELTYPE', y='CO2EMISSIONS', data=df, palette='rainbow')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.violinplot(x='VEHICLECLASS', y='CO2EMISSIONS', data=df, palette='rainbow')
plt.ylabel('CO2 Emissions')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
for col in quantitative_cols:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(x=col, data=df, fill=True, palette='viridis')
    plt.title(f'Density Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```python
for col in quantitative_cols:
    sns.histplot(x=col, data=df, fill=True, color='turquoise')
    plt.title(f'Density Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```python
print('Skewness for quantitative features:')
for col in quantitative_cols:
    print(f'{col}: {df[col].skew():.2f}')
    sns.kdeplot(x=col, data=df, fill=True, color='cyan')
    plt.title(f'Density Plot of {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```
